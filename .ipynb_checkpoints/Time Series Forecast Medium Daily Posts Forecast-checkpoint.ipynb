# Time Series Forecast: Medium daily posts forecast

# Topic

 A time series is simply a series of data points ordered in time. In a time series, time is often the independent variable and the goal is usually to make a forecast for the future.

In this notebook I chose a dataset which contains timestamps and urls of Medium publications over the last several years, I will make the necessary transformation and apply two kinds of algorithms: ETS (error,trend and seasonality), and ARIMA algorithms (Autoregressive, Integrated, Moving Average), then I'm going to compare the results and pick the model that best fits the data.

# Objectives

- Decompose and analyse the time series
- Make forecasts regarding daily Medium posts

## ETS Models

To be able to do Error Trend Seasonality analysis, I sould first plot the decomposition of the TS into its three components.

```python
decomposition = sm.tsa.seasonal_decompose(daily)
```

```python
fig, axes = plt.subplots(4, 1, figsize=(12, 10))
daily.plot(title='Original', ax=axes[0])
decomposition.trend.plot(title='Trend', ax=axes[1])
decomposition.seasonal.plot(title='Seasonality', ax=axes[2])
decomposition.resid.plot(title='error', ax=axes[3])
```

The decomposition of the time series shows a linear uptrend, the seasonality doesn't change in magnitude, is weekly and looks constant over time and the error also seems constant. Therefore for the trend an additive method should be used, for the seasonality and error also an additive method is the way to go.

## Making Predictions

## Simple Exponentioanl Smoothing

The simple exponential Smoothing method is normally used for Time Series with no trend and no seasonality. Since this dataset at hand has both, I going to assume that SES will not do great with it. That's why I'll use it as my baseline model. To make it work better, I'm going to set its alpha to a high number in order for the model to give higher weights to the latest observations since it looks form the plot that the sooner is the observation, the higher the number of posts.

```python
ses = SimpleExpSmoothing(train).fit(smoothing_level=0.9,optimized=False)
```

```python
ses.summary()
```

The table above is the summary of the Simple Exponential smoothing, it shows a lot of information about the model, most important of which is the AIC model which determines how the model fits to the data. Now this number 5099.45 cannot tell us anything by itself, it has to be compared with that of other models to help choose the best one.

```python
ses_pred = ses.forecast(len(test))
ses_rmse = mean_squared_error(test, ses_pred, squared=False)
```

```python
ses_mae = mean_absolute_error(test, ses_pred)
ses_mae
```

```python
# function to plot the forecast
def plot_forecast(forecast, model, rmse):
    plt.figure(figsize=(12,4))
    plt.plot(train, label='Training')
    plt.plot(test, label='Actual')
    plt.plot(forecast, label='Forecast')
    plt.xlabel('Datetime')
    plt.ylabel('Posts Published')
    plt.title(f'{model} (RMSE: {round(rmse, 2)})')
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
plot_forecast(ses_pred, "Simple Exponential Smoothing", ses_rmse)
```

Just like expected, the SES predicts a straight line that is not going with the trend of the data. The RMSE is also high 100.09.

```python
l = list(ses_pred)
t = test.url
t = list(t)
```

```python
ses_res = [t[i] - l[i] for i in range(len(l))]
```

```python
plot_acf(ses_res, lags=40)
```

The Autocorrelation Plot of the residuals of teh Simple Exponential Smoothing shows strong correlations at lags 1 and 2 suggesting that there is still a lot of information in the data that hasn't been accountes for by the SES model, which is very normal considering it is only the baseline model.

## Holt's Linear Trend

The next condidate would be probably Holt's Linear trend, but this model is used when there is trend but no seasonality in the data. I'm going to try it nevertheless to see how much better than SES it does.

The forecast made by Holt's Linear model alignes better with the actual values and has a smaller RMSE. Let's remember that this model though accounts only on for the trend but not the seasonality.

## Holt Winters seasonal method

The next model I'm going to try is Holt Winters Seasonal Method. This method accounts for both the trend and seasonality in the times series, it also allows to adapt both the additive and multiplicative method for either the trend or seasonality. From the TS decomposition, it seems that the trend grows linearly in the train set but exponentially on the test set, therefore a multiplicative method will be aplied for it, and shows a seasonality that doesn't grow in magnitude and is fairly constant, so an additive method will be applied to the seasonlity.

In my model, I will set the seasonal periods to 7 because the data is collected daily and has a weekly seasonality. I'm going to set trend and seasonal to additive. I will also set the use boxcox to true in order to transform the data into a normal distribution to be better handeled by the model.

```python
hwm = ExponentialSmoothing(train, seasonal_periods=7, trend='mul', seasonal='add',use_boxcox=True).fit()
```

```python
hwm.summary()
```

And for holt Winter's exponential model which accounts for both the seasonality and trend in the data, the AIC is lower and better than previous models.

```python
hwm_pred = hwm.forecast(len(test))
```

```python
hwm_rmse =  mean_squared_error(test, hwm_pred, squared=False)
```

```python
hwm_mae = mean_absolute_error(test, hwm_pred)
hwm_mae
```

```python
plot_forecast(hwm_pred, "Holt Winters Seasonal Method", hwm_rmse)
```

The forecast made by Holt Winters Seasonal Method are the best sofar, we see that the forecast aligns better with the actual values, and the RMSE is reduced further. This makes sense because this model is the one that takes into consideration the trend and seasonality of the data.

```python
l = list(hwm_pred)
hwm_res = [t[i] - l[i] for i in range(len(l))]
plot_acf(hwm_res, lags=40)
```

The residuals ACF plot however shows a strong correlation for both lags 1 and 2 meaning that a lot of information is still unexpoited by the model.

## ARIMA Models : SARIMA

In order to use an ARIMA model, the time series should be stationary. Now we know from the decomposition plot that this series has both and uptrend and a seasonality but let's confirm that more with the  Augmented Dickey-Fuller or ADF test. Now the ADF test return a p value which has to be less than .05 for the data to be stationary and if it is above the threshold that means that tha data is not stationary.

### Checking for Stationarity

```python
stat = adfuller(daily)
```

```python
print('p-value: %f' % stat[1])
```

### Making forecasts

Now it's time to build a model, I'm going to use Seasonal ARIMA or SARIMA to account for both the seasonality and trend in the data, the terms will be ARIMA(1,1,0),(0,1,1)7 because:
- non seasonal AR is 1 because there is positive correlation in the first lag of the ACF and PACF drops quickly after.
- non seasonal differencing is at 1 because I applied the first differencing on the data
- non seasonal MA is at 0 because there is no indicator that we need MA terms
- seasonal AR is at 0 because there is no indication that we need AR terms
- seasonal differencing is at 1 because I used seasonal difference as well 
- seasonal MA is at 1 because the first seasonal lag i.e lag number 7 is negative.
- 7 is the number of days in the seasonal cycle of the the time series

```python
sarima = sm.tsa.statespace.SARIMAX(train, order=(1,1,0), seasonal_order=(0,1,1,7))
```

```python
results = sarima.fit()
```

```python
print(results.summary() )
```
